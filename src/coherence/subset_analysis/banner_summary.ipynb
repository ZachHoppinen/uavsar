```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import plotly.express as px
import xarray as xr
```

```python
sub = xr.open_dataset('../data/netcdfs/banner_lowman_05208_21019-019_21021-007_0006d.nc')
```

```python
d1 = pd.to_datetime(sub.attrs['start time of acquisition for pass 1'])
d2 = pd.to_datetime(sub.attrs['start time of acquisition for pass 2'])
```

```python
wx_fp = '../data/wx_station/banner/'
assert d1.year == d2.year, "Years don't match"
for csv_fp in os.listdir(wx_fp):
    if str(d1.year) in csv_fp:
        df = pd.read_csv(os.path.join(wx_fp, csv_fp), skiprows = 3)
        df['date_time'] = pd.to_datetime(df.Date +' '+  df.Time)
        df = df.set_index('date_time')
        df = df.drop(['Date','Time'],axis = 1)
```

```python
win = df[(df.index > pd.to_datetime(d1.tz_localize(None).floor('H'))) & (df.index < pd.to_datetime(d2.tz_localize(None).floor('H')))]
```

```python
fig = px.line(win, x=win.index, y='PREC.I-1 (in) ',
              labels={
                     "PREC.I-1 (in) ": "Precipitation (in)",
                     "date_time": "Date"
                 }
             )
fig.show()
```
