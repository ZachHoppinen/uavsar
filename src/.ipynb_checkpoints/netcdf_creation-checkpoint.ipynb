```python
class Image:
    def __init__(self, fp, dtype, pol, real_im):
        self.fp = fp
        self.dtype = dtype
        self.pol = pol
        self.comp = real_im
    array = None
    top = None
    bottom = None
    left = None
    right = None
    crs = None
    transform = None
    nodata = None
j = 0
img_list = []
        
for im in glob(loc_dir + '/*.tif'):
    print(im)
    if 'csv' in im:
        desc = pd.read_csv(join(loc_dir, im))
    if 'tif' in im:
        fp = join(loc_dir, im)
        if len(basename(im).split('.')) == 3:
            real_im = 'real'
            dtype = basename(im).split('.')[-2]
            pol = np.nan

        else:
            real_im = basename(im).split('.')[-2]
            dtype = basename(im).split('.')[-3]
            pol = basename(im).split('.')[0].split('_')[-2][4:]
        
        img = Image(fp, dtype, pol, real_im)
        if 'inc' in fp:
            with rio.open(fp) as src:
                arr = src.read(
                out_shape=(
                    src.count,
                    target_shape[0],
                    target_shape[1]
                ),
                resampling=Resampling.bilinear
            )[0]
                img.array = arr
                print(arr.shape)
                print(np.mean(arr))
        
        else:
            with rio.open(fp) as src:
                img.crs = src.crs
                img.transform = src.transform
                boundary = src.bounds
                my_proj = pyproj.Proj(str(src.crs).lower())#your data crs
                left_bottom = my_proj(boundary[0],boundary[1], inverse=True)
                right_top = my_proj(boundary[2],boundary[3], inverse=True)
                img.left = left_bottom[0]
                img.right = right_top[0]
                img.top = right_top[1]
                img.bottom = left_bottom[1]
                arr = src.read(1)
                target_shape = arr.shape
                img.array = arr
                img.nodata = src.nodata
                print(np.nanmin(arr))
        if j == 3:
            pass
        j+=1
        print(sys.getsizeof(img.array))
        img_list.append(img)
```
